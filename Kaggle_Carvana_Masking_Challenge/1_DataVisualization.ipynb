```python
os.getcwd()
```

```python
DATA_PATH = '../DeepLearningDataset/Kaggle_Carvana_Masking_Challenge/dataset'

TRAIN_DATA_PATH = os.path.join(DATA_PATH,'/train')
TRAIN_MASKS_PATH = os.path.join(DATA_PATH,'train_masks')

METADATA_PATH = os.path.join(DATA_PATH,'metadata.csv')
metadata_df = pd.read_csv(METADATA_PATH,header='infer')

TRAIN_MASK_FILE = os.path.join(DATA_PATH,'train_masks.csv')
train_mask_df = pd.read_csv(TRAIN_MASK_FILE,header='infer')

TEST_DATA_PATH = os.path.join(DATA_PATH,'/test')


SUBMISSION_FILE = os.path.join(DATA_PATH,'sample_submission.csv')
submission_df = pd.read_csv(SUBMISSION_FILE,header='infer')
```
